# Image Color Space

```python
import numpy as np 
import matplotlib.pyplot as plt 
from skimage import io, data, img_as_ubyte, img_as_float
from skimage.color import rgb2hsv, rgb2lab, hsv2rgb
from skimage.transform import rotate
from skimage.exposure import equalize_hist
```

## 1. RGB Color Image

```python
img = data.astronaut()

plt.imshow(img)
plt.axis('off')
plt.show()
```

## 2. isolate single color channel

```python
def rgb_isolate_channel(im, channel):
    """ 
    Isolate a single color channel in an RGB image.
    
    Parameters:
        im (ndarray): Input RGB image.
        channel (str): Color channel to isolate ('R', 'G', or 'B').
    Returns:
        ndarray: Image with only the specified channel.
    """
    imc = im.copy()
    for c in range(im.shape[2]):
        if c != channel:
            imc[:, :, c] = np.zeros_like(im[:, :, c])
    return imc

img = data.astronaut()
img_r = rgb_isolate_channel(img, 0)  # Isolate Red channel
img_g = rgb_isolate_channel(img, 1)  # Isolate Green channel
img_b = rgb_isolate_channel(img, 2)  # Isolate Blue channel

fig = plt.figure()
ax = fig.add_subplot(1, 3, 1)
ax.imshow(img_r)
ax.set_title('Red Channel')
ax.axis('off')

ax = fig.add_subplot(1, 3, 2)
ax.imshow(img_g)
ax.set_title('Green Channel')
ax.axis('off')

ax = fig.add_subplot(1, 3, 3)
ax.imshow(img_b)
ax.set_title('Blue Channel')
ax.axis('off')
```

## 3. Isolate single channel

```python
img = data.astronaut()
img_r = img[:,:,0]
img_g = img[:,:,1]
img_b = img[:,:,2]
fig = plt.figure()

ax = fig.add_subplot(1, 3, 1)
p = plt.imshow(img_r, cmap='gray')
ax.set_title('Red Channel in GS')
ax.axis('off')
c = plt.colorbar(orientation='horizontal')
plt.clim(0, 255)

ax = fig.add_subplot(1, 3, 2)
p = plt.imshow(img_g, cmap='gray')
c = plt.colorbar(orientation='horizontal')
plt.clim(0, 255)
ax.set_title('Green Channel in GS')
ax.axis('off')

ax = fig.add_subplot(1, 3, 3)
p = plt.imshow(img_b, cmap='gray')
c = plt.colorbar(orientation='horizontal')
plt.clim(0, 255)
ax.set_title('Blue Channel in GS')
ax.axis('off')
```

## 4. RGB to HSV


```python
img = data.astronaut()
img_hsv = rgb2hsv(img)
img_hsv_h = img_hsv[:,:,0]
img_hsv_s = img_hsv[:,:,1]
img_hsv_v = img_hsv[:,:,2]

fig = plt.figure()
ax = fig.add_subplot(1, 3, 1)
p = plt.imshow(img_hsv_h, cmap='gray')
c = plt.colorbar(orientation='horizontal')
ax.set_title('Hue Channel')
ax.axis('off')


ax = fig.add_subplot(1, 3, 2)
p = plt.imshow(img_hsv_s, cmap='gray')
c = plt.colorbar(orientation='horizontal')
ax.set_title('Saturation Channel')
ax.axis('off')

ax = fig.add_subplot(1, 3, 3)
p = plt.imshow(img_hsv_v, cmap='gray')
c = plt.colorbar(orientation='horizontal')
ax.set_title('Value Channel')
ax.axis('off')
```

## 5. RGB to CIE LAB

```python
img = data.astronaut()
img_lab = rgb2lab(img)
img_lab_l = img_lab[:,:,0]
img_lab_a = img_lab[:,:,1]
img_lab_b = img_lab[:,:,2]

fig = plt.figure()
ax = fig.add_subplot(1, 3, 1)
p = plt.imshow(img_lab_l, cmap='gray')
c = plt.colorbar(orientation='horizontal')
ax.set_title('L* Channel')
ax.axis('off')

ax = fig.add_subplot(1, 3, 2)
p = plt.imshow(img_lab_a, cmap='gray')
c = plt.colorbar(orientation='horizontal')
ax.set_title('a* Channel')
ax.axis('off')

ax = fig.add_subplot(1, 3, 3)
p = plt.imshow(img_lab_b, cmap='gray')
c = plt.colorbar(orientation='horizontal')
ax.set_title('b* Channel')
ax.axis('off')
```

## 6. Color Mapping

```python
img = data.astronaut()
img_gray = img[:,:,0]
cm = plt.get_cmap('hot')
img_idx1 = cm(img_gray)
cm = plt.get_cmap('jet')
img_idx2 = cm(img_gray)
cm = plt.get_cmap('hsv')
img_idx3 = cm(img_gray)
cm = plt.get_cmap('terrain')
img_idx4 = cm(img_gray)
cm = plt.get_cmap('bone')
img_idx5 = cm(img_gray)
cm = plt.get_cmap('copper')
img_idx6 = cm(img_gray)
fig = plt.figure()
ax = fig.add_subplot(2, 3, 1)
plt.imshow(img_idx1)
ax.axis('off')
ax = fig.add_subplot(2, 3, 2)
plt.imshow(img_idx2)
ax.axis('off')
ax = fig.add_subplot(2, 3, 3)
plt.imshow(img_idx3)
ax.axis('off')
ax = fig.add_subplot(2, 3, 4)
plt.imshow(img_idx4)
ax.axis('off')
ax = fig.add_subplot(2, 3, 5)
plt.imshow(img_idx5)
ax.axis('off')
ax = fig.add_subplot(2, 3, 6)
plt.imshow(img_idx6)
ax.axis('off')
plt.show()
```
